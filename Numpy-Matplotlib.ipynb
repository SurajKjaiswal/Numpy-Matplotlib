```python
import matplotlib.pyplot as plt # type: ignore
import numpy as np
categories = ['A', 'B', 'C', 'D', 'E']
values1 = [5, 7, 3, 8, 4]  # Group 1
values2 = [6, 5, 7, 6, 5]  # Group 2
bar_width=0.4
x=np.arange(len(categories))
plt.bar(x-bar_width/2,values1,width=bar_width,color='red',label='Group1',alpha=0.5)
plt.bar(x+bar_width/2,values2,width=bar_width,color='blue',label='Group2',alpha=0.5)
plt.plot(x-bar_width/2,values1,color='red')
plt.plot(x+bar_width/2,values2,color='blue')
#plt.xlabel("Categories")
plt.xticks(x,categories)
plt.ylabel("Values")
plt.title("Side By Side Bar Chart with Multiple group")
plt.legend()
plt.show()
```

```python
import numpy as np # type: ignore
import matplotlib.pyplot as plt # type: ignore
# Sample student data
students = ['Jhon', 'Smith', 'Marry', 'Rose', 'Devid']
math_scores = [85, 92, 78, 88, 90]
science_scores = [76, 88, 92, 80, 78]
english_scores = [88, 96, 84, 90, 73]
cs_scores = [70, 85, 95, 85, 75]
fig, axs = plt.subplots(2, 2, figsize=(20, 8))
# Ploting Math Scores
axs[0][0].bar(students, math_scores, color='blue')
axs[0][0].set_title('Math Scores')
axs[0][0].set_xlabel('Students')
axs[0][0].set_ylabel('Score')
# Ploting Science Scores
axs[0][1].bar(students, science_scores, color='orange')
axs[0][1].set_title('Science Scores')
axs[0][1].set_xlabel('Students')
axs[0][1].set_ylabel('Score')
# Ploting English Scores
axs[1][0].bar(students, english_scores, color='grey')
axs[1][0].set_title('English Scores')
axs[1][0].set_xlabel('Students')
axs[1][0].set_ylabel('Score')
# Ploting Computer Scores
axs[1][1].bar(students, cs_scores, color='green')
axs[1][1].set_title('Computer Scores')
axs[1][1].set_xlabel('Students')
axs[1][1].set_ylabel('Score')
# Adjust spacing between subplots
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt # type: ignore
product_prices = [24.99, 34.99, 49.99, 64.99, 39.99, 54.99, 79.99, 99.99, 29.99, 44.99, 59.99, 69.99, 84.99, 109.99, 119.99, 89.99, 74.99, 124.99, 69.99, 54.99]
bins = [20, 40, 60, 80, 100, 120, 140]  
# Create Histogram
plt.hist(product_prices, bins=bins, color='pink', edgecolor='black', alpha=0.7)
plt.xlabel('Price Range ', fontsize=12, fontweight='bold')
plt.ylabel('Number of Products', fontsize=12, fontweight='bold')
plt.title('Distribution of Product Prices in a Retail Store', fontsize=14, fontweight='bold')
plt.xticks(bins)  
plt.show()
```

#Conclusion
Product lies between (40 - 80) is higher
Product lies between (20 - 40) is average
After 80 the number of product is less

```python
import numpy as np
import matplotlib.pyplot as plt

square_footage = np.array([1200, 1400, 1600, 1800, 2000, 2200, 2400, 2600, 2800, 3000])
selling_prices = np.array([250, 290, 315, 380, 410, 450, 500, 525, 570, 610])  # in $1000s
plt.figure(figsize=(8, 10))
plt.scatter(square_footage, selling_prices, color='blue', marker='o', label='Data Points')
plt.plot(square_footage,selling_prices)
plt.xlabel('House Size (Square Footage)', fontsize=12, fontweight='bold')
plt.ylabel('Selling Price ', fontsize=12, fontweight='bold')
plt.title('House Size vs. Selling Price', fontsize=14, fontweight='bold')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data: Sales for 12 months
months = np.arange(1, 13)
electronics_sales = np.array([25000, 28000, 31000, 27000, 30000, 32000, 35000, 36000, 38000, 39000, 41000, 42000])
clothing_sales = np.array([15000, 16000, 17000, 18000, 19000, 20000, 21000, 22000,
23000, 24000, 25000, 26000])
home_garden_sales = np.array([18000, 19000, 20000, 21000, 22000, 23000, 24000,
25000, 26000, 27000, 28000, 29000])
sports_outdoors_sales = np.array([12000, 13000, 14000, 15000, 16000, 17000, 18000,
19000, 20000, 21000, 22000, 23000])

plt.plot(months, electronics_sales, marker='o', linestyle='-', color='blue', label='Electronics')
plt.plot(months, clothing_sales, marker='s', linestyle='-', color='red', label='Clothing')
plt.plot(months, home_garden_sales, marker='^', linestyle='-', color='green', label='Home & Garden')
plt.plot(months, sports_outdoors_sales, marker='d', linestyle='-', color='orange', label='Sports & Outdoors')
plt.xlabel('Months', fontsize=12, fontweight='bold')
plt.ylabel('Sales', fontsize=12, fontweight='bold')
plt.title('Sales Performance by Product Category', fontsize=14, fontweight='bold')
plt.legend()
plt.show()
```

As we can see that the sales of Electronic is fluctuating during the middle months but there is highest increasing sales at last months.
And the sales of Clothing,Home & garden, Sports & Outdoors is increasing slowly but increasing constently every months.

```python
import numpy as np
import matplotlib.pyplot as plt

# Data: Sales for 12 months
months = np.arange(1, 13)
electronics_sales = np.array([25000, 28000, 31000, 27000, 30000, 32000, 35000, 36000, 38000, 39000, 41000, 42000])
clothing_sales = np.array([15000, 16000, 17000, 18000, 19000, 20000, 21000, 22000,
23000, 24000, 25000, 26000])
home_garden_sales = np.array([18000, 19000, 20000, 21000, 22000, 23000, 24000,
25000, 26000, 27000, 28000, 29000])
sports_outdoors_sales = np.array([12000, 13000, 14000, 15000, 16000, 17000, 18000,
19000, 20000, 21000, 22000, 23000])
fig, axs = plt.subplots(2, 2, figsize=(20, 12))
#Ploting electronics_sales
axs[0][0].bar(months, electronics_sales, color='blue')
axs[0][0].set_title('electronics_sales')
axs[0][0].set_xlabel('months')
axs[0][0].set_ylabel('electronics_sales')
# Ploting clothing_sales
axs[0][1].bar(months, clothing_sales, color='orange')
axs[0][1].set_title('clothing_sales ')
axs[0][1].set_xlabel('months')
axs[0][1].set_ylabel('clothing_sales')
# Ploting home_garden_sales
axs[1][0].bar(months, home_garden_sales, color='grey')
axs[1][0].set_title('home_garden_sales')
axs[1][0].set_xlabel('months')
axs[1][0].set_ylabel('home_garden_sales')
# Ploting Computer Scores
axs[1][1].bar(months, sports_outdoors_sales, color='green')
axs[1][1].set_title('sports_outdoors_sales')
axs[1][1].set_xlabel('months')
axs[1][1].set_ylabel('sports_outdoors_sales')
# Adjust spacing between subplots
plt.tight_layout()
plt.show()
```
